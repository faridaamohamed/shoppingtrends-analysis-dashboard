```python
bins = [10, 20, 30, 40, 50, 60, 70]
labels = ['10-20', '21-30', '31-40', '41-50', '51-60', '61-70']
df['Age_Range'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

cross2 = pd.crosstab(df["Age_Range"], df["Preferred_Payment_Method"])

preferred_by_age_range = cross2.idxmax(axis=1) 
preferred_by_age_range.name = "Top_Preferred_Payment_Method"

result = pd.concat([cross2, preferred_by_age_range], axis=1)

result
```

```python
special_customer = df.groupby(["Location", "Customer_ID"])["Purchase_Amount_USD"].sum().reset_index()

special_customer_sorted = special_customer.sort_values(["Location", "Purchase_Amount_USD"], ascending=[True, False])

top_10_per_location = special_customer_sorted.groupby("Location").head(10)

top_10_per_location
```

```python
def get_top_customers_by_location(df, location, top_n=10):
    special_customer = df.groupby(["Location", "Customer_ID"])["Purchase_Amount_USD"].sum().reset_index()
    
    filtered = special_customer[special_customer["Location"] == location]

    sorted_filtered = filtered.sort_values("Purchase_Amount_USD", ascending=False)
    top_customers = sorted_filtered.head(top_n)

    return top_customers
```

```python
location_input = input("enter the location: ")  
result = get_top_customers_by_location(df, location_input)
print(result)
```
